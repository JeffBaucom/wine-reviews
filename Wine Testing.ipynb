```python
import numpy as np
import csv
import sys
import pandas as pd

#reload(sys)
#sys.setdefaultencoding('utf8')
wine_data = pd.read_csv('winemag-data-130k-v2.csv', sep=',', encoding='utf-8')
print wine_data.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer()
tf_result = vec.fit_transform(wine_data.description)
feature_names = vec.get_feature_names()
for i in range(10):
    doc = i
    feature_index = tf_result[doc, :].nonzero()[1]
    tfidf_scores = zip(feature_index, [tf_result[doc, x] for x in feature_index])
    for w, s in [(feature_names[i], s) for (i, s) in tfidf_scores]:
      print w, s
```
